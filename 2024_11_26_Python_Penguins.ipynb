```python
def p_value_to_words(p_value):
  p_value_rounded = str(round(p_value, 3))

  if p_value >= 0.05:
    print("We cannot reject null hypothesis as p value is " + p_value_rounded)
  elif p_value < 0.001:
    print("*** p value s,aller than threshold, p value is " + p_value_rounded)
  elif p_value < 0.05:
    print("* p value is smaller than critical, p value is " + p_value_rounded)
  else:
    print("p value defies maths. Don't know what to do " + p_value_rounded)

p_value_to_words(0.05)
p_value_to_words(0.00005)
p_value_to_words(0.01)
```
